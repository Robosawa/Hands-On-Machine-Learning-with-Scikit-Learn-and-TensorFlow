```
!pip install janome
all_text = "When this add-in is used,Can read and make changes to your documenCan send data over the Internet add add add read read read read"
from janome.tokenizer import Tokenizer
tok = Tokenizer()
tokens = tok.tokenize(all_text)
words = ""

for token in tokens:
    if token.part_of_speech.split(',')[0] in ['名詞', '動詞', '形容詞', '副詞']:
        words = words + " " + token.base_form

from wordcloud import WordCloud
wordcloud = WordCloud(background_color="white"
                    ,width=800
                    ,height=500).generate(words)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show
```
